```python
# Conversion des étiquettes catégorielles en valeurs binaires:
# - "G" (Goodware) devient 0
# - "R" (Ransomware) devient 1
# Cette transformation est nécessaire pour les algorithmes d'apprentissage automatique qui requièrent des entrées numériques
df['Sample_Type'] = df['Sample_Type'].map({"G" : 0, "R" : 1}).astype(int)

# Afficher la distribution des classes après conversion
print("Distribution des classes après conversion:")
print(df['Sample_Type'].value_counts())
print(f"Pourcentage: {df['Sample_Type'].value_counts(normalize=True).round(4) * 100}")
```

## Features Engineering

```python
# Définir les listes d'API de bas niveau
colonnes_reseau = ['bind', 'connect', 'socket', 'closesocket', 'WSASocketW', 'WSAStartup',
                  'WSARecv', 'WSASend', 'getaddrinfo', 'GetAddrInfoW', 'sendto', 'setsockopt',
                  'InternetOpenUrlA', 'InternetOpenA', 'InternetConnectA', 'HttpOpenRequestA',
                  'InternetReadFile', 'shutdown', 'unique_domains_contacted']

colonnes_fichier = ['CreateFileW', 'WriteFile', 'ReadFile', 'DeleteFileW', 'MoveFileWithProgressW',
                   'CopyFileW', 'CreateDirectoryW', 'FindFirstFileExW', 'SetFileAttributesW',
                   'SetFilePointer', 'SetFilePointerEx', 'SetFileTime', 'SetEndOfFile', 'GetFileSize',
                   'NtCreateFile', 'NtReadFile', 'NtWriteFile', 'NtDeleteFile', 'dropped_files_count']

colonnes_processus = ['CreateProcessInternalW', 'CreateRemoteThreadEx', 'CreateThread', 'OpenProcess',
                     'WriteProcessMemory', 'ReadProcessMemory', 'VirtualAllocEx', 'VirtualProtectEx',
                     'NtOpenProcess', 'NtCreateThreadEx', 'CreateToolhelp32Snapshot', 'Thread32Next',
                     'Process32Next']

# Pour le registre, on considére toutes les colonnes commençant par 'HKEY'
colonnes_registre = [col for col in df_sans_doublons.columns if col.startswith('HKEY')]

reseau_existant = [col for col in colonnes_reseau if col in df_sans_doublons.columns]
fichier_existant = [col for col in colonnes_fichier if col in df_sans_doublons.columns]
processus_existant = [col for col in colonnes_processus if col in df_sans_doublons.columns]

# Créer les nouvelles caractéristiques comportementales de haut niveau en additionnant les colonnes correspondantes
df_sans_doublons['activite_reseau'] = df_sans_doublons[reseau_existant].sum(axis=1)
df_sans_doublons['activite_fichier'] = df_sans_doublons[fichier_existant].sum(axis=1)
df_sans_doublons['activite_processus'] = df_sans_doublons[processus_existant].sum(axis=1)
df_sans_doublons['activite_registre'] = df_sans_doublons[colonnes_registre].sum(axis=1)

# Afficher un échantillon des nouvelles caractéristiques
df_sans_doublons[['activite_reseau', 'activite_fichier', 'activite_processus', 'activite_registre']].head()
```

```python
# 1. Utilise des mécanismes de persistance - vérifie les clés de registre d'exécution ou paramètres de démarrage
cles_persistance = [
    'HKEY_CURRENT_USER\\Software\\Microsoft\\Windows\\CurrentVersion\\Run',
    'HKEY_LOCAL_MACHINE\\Software\\Microsoft\\Windows\\CurrentVersion\\Run',
    'HKEY_CURRENT_USER\\Software\\Microsoft\\Windows\\CurrentVersion\\Policies\\Explorer\\',
    'HKEY_LOCAL_MACHINE\\SOFTWARE\\Wow6432Node\\Microsoft\\Windows\\CurrentVersion\\'
]
df_sans_doublons['utilise_persistance'] = df_sans_doublons[cles_persistance].sum(axis=1) > 0

# 2. Modifie les fichiers système - recherche des opérations sur les fichiers et les répertoires système
indicateurs_fichiers_systeme = [
    'WriteFile', 'NtWriteFile', 'MoveFileWithProgressW', 
    'SetFileAttributesW', 'CopyFileW', 'DeleteFileW'
]
indicateurs_fichiers_systeme = [col for col in indicateurs_fichiers_systeme if col in df_sans_doublons.columns]
df_sans_doublons['modifie_fichiers_systeme'] = (df_sans_doublons[indicateurs_fichiers_systeme].sum(axis=1) > 0) & \
                                              (df_sans_doublons['activite_fichier'] > 5)  # Seuil plus élevé pour l'activité système

# 3. Présence réseau - tous les appels liés au réseau
df_sans_doublons['presence_reseau'] = df_sans_doublons['activite_reseau'] > 0

# 4. Chiffre les fichiers - présence de fonctions de chiffrement plus opérations sur les fichiers
indicateurs_chiffrement = [
    'CryptEncrypt', 'CryptCreateHash', 'CryptHashData', 'CryptExportKey'
]
indicateurs_chiffrement = [col for col in indicateurs_chiffrement if col in df_sans_doublons.columns]
operations_fichiers = ['WriteFile', 'ReadFile', 'SetFilePointer', 'SetFileTime']
operations_fichiers = [col for col in operations_fichiers if col in df_sans_doublons.columns]

df_sans_doublons['chiffre_fichiers'] = (df_sans_doublons[indicateurs_chiffrement].sum(axis=1) > 0) & \
                                    (df_sans_doublons[operations_fichiers].sum(axis=1) > 5)

# 5. Manipulation élevée de processus/mémoire (courante dans les logiciels malveillants)
manipulation_processus = [
    'CreateRemoteThreadEx', 'WriteProcessMemory', 'ReadProcessMemory', 
    'VirtualAllocEx', 'VirtualProtectEx'
]
manipulation_processus = [col for col in manipulation_processus if col in df_sans_doublons.columns]
df_sans_doublons['manipule_processus'] = df_sans_doublons[manipulation_processus].sum(axis=1) > 0

# Afficher la distribution de ces indicateurs par type d'échantillon
print("Indicateurs de comportement suspect par type d'échantillon (0=Goodware, 1=Ransomware):")
for indicateur in ['utilise_persistance', 'modifie_fichiers_systeme', 'presence_reseau', 
             'chiffre_fichiers', 'manipule_processus']:
    tableau_croise = pd.crosstab(df_sans_doublons['Sample_Type'], df_sans_doublons[indicateur])
    pourcentage = pd.crosstab(df_sans_doublons['Sample_Type'], df_sans_doublons[indicateur], 
                            normalize='index') * 100
    
    print(f"\n{indicateur.replace('_', ' ').title()}:")
    print(f"Comptages bruts:\n{tableau_croise}")
    print(f"Pourcentage dans le groupe:\n{pourcentage.round(2)}")

# Créer un score de suspicion combiné
df_sans_doublons['score_suspicion'] = df_sans_doublons[['utilise_persistance', 'modifie_fichiers_systeme', 
                                                      'presence_reseau', 'chiffre_fichiers', 
                                                      'manipule_processus']].sum(axis=1)

# Afficher la distribution du score de suspicion
print("Distribution du score de suspicion:")
distribution_score = df_sans_doublons.groupby(['Sample_Type', 'score_suspicion']).size().unstack()
print(distribution_score)
```

```python
# Calculer le nombre d'appels au registre et le total des appels API
nombre_appels_registre = df_sans_doublons[colonnes_registre].sum(axis=1)
nombre_appels_api_total = df_sans_doublons.drop(columns=['Sample_Type'] + colonnes_registre).sum(axis=1)

# Calculer le ratio registre/API
ratio_registre_api = nombre_appels_registre / (nombre_appels_api_total+1)
df_sans_doublons['ratio_registre_api'] = ratio_registre_api

# Calculer le nombre d'opérations sur fichiers et sur processus
nombre_ops_fichier = df_sans_doublons[fichier_existant].sum(axis=1)
nombre_ops_processus = df_sans_doublons[processus_existant].sum(axis=1)

# Calculer le ratio fichier/processus
ratio_fichier_processus = nombre_ops_fichier / (nombre_ops_processus+1)
df_sans_doublons['ratio_fichier_processus'] = ratio_fichier_processus

# Afficher les premières lignes avec les nouvelles caractéristiques
print(df_sans_doublons[['ratio_registre_api', 'ratio_fichier_processus']].head())
```

```python
import numpy as np

# Créer une copie du DataFrame pour la transformation
df_indicateurs_log = df_sans_doublons.copy()

# Identifier toutes les colonnes numériques à transformer (exclure Sample_Type)
colonnes_a_transformer = [col for col in df_sans_doublons.columns if col != 'Sample_Type']

# Appliquer la transformation logarithmique log(x+1) à toutes les colonnes numériques
for colonne in colonnes_a_transformer:
    # Vérifier si la colonne est numérique et pas booléenne
    if df_indicateurs_log[colonne].dtype in ['int64', 'float64'] and not (df_indicateurs_log[colonne].isin([0, 1]).all()):
        df_indicateurs_log[colonne] = np.log1p(df_indicateurs_log[colonne])

# Enregistrer les données transformées dans un fichier CSV
df_indicateurs_log.to_csv('dataLogTransforme.csv', index=False)

print(f"Données avec indicateurs transformés par logarithme enregistrées dans 'dataLogTransforme.csv'")
print(f"Dimensions des données transformées: {df_indicateurs_log.shape}")

# Afficher un aperçu des données transformées
df_indicateurs_log[['activite_reseau', 'activite_fichier', 'activite_processus', 'activite_registre', 
                   'ratio_registre_api', 'ratio_fichier_processus', 'score_suspicion']].head()
```

```python
# Créer une copie du DataFrame pour l'encodage binaire
df_binaire = df_sans_doublons.copy()

# Conserver la colonne Sample_Type qui est déjà codée en binaire (0 ou 1)
target_column = 'Sample_Type'

# Pour toutes les autres colonnes, convertir en binaire (0 ou 1)
for colonne in df_binaire.columns:
    if colonne != target_column:
        # Convertir les booléens en entiers (0 ou 1)
        if df_binaire[colonne].dtype == 'bool':
            df_binaire[colonne] = df_binaire[colonne].astype(int)
        # Convertir les valeurs numériques en binaire (0 si valeur=0, 1 sinon)
        elif pd.api.types.is_numeric_dtype(df_binaire[colonne]):
            df_binaire[colonne] = (df_binaire[colonne] > 0).astype(int)

# Enregistrer le DataFrame avec l'encodage binaire
df_binaire.to_csv('dataBinaire.csv', index=False)

# Afficher les dimensions et un aperçu du DataFrame encodé en binaire
print(f"Dimensions du DataFrame encodé en binaire: {df_binaire.shape}")
print("\nAperçu du DataFrame encodé en binaire:")
df_binaire.head()
```
